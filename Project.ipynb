```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("D:\\ITI\\Power BI Development\\ORG. Internships\\NeuronetiX Internship\\Session 3 Seaborn, Scikit-Learn\\Seaborn, Scikit-Learn Task\\data.csv")

df.dropna(inplace = True)
df.fillna(10000 , inplace = True)
# S = df["Calories"].mode()[0]
# df["Calories"].fillna(S,inplace = True)
df.at[26, 'Date'] = "'2020/12/26'" 
# print(df.duplicated())
df.drop_duplicates(inplace = True)
print(df)

```

```python
# 1. Line Plot of Calories burned over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Calories', data=df, marker='o', color='blue')
plt.title('Calories Burned Over Time')
plt.xlabel('Date')
plt.ylabel('Calories')
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Scatter Plot of Pulse vs. Calories
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Pulse', y='Calories', data=df, color='green')
plt.title('Pulse vs. Calories Burned')
plt.xlabel('Pulse')
plt.ylabel('Calories')
plt.show()
```

```python
# 3. Histogram of Duration
plt.figure(figsize=(10, 6))
sns.histplot(df['Duration'], bins=10, kde=True, color='orange')
plt.title('Distribution of Exercise Duration')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.show()
```

```python
# 4. Boxplot of Maxpulse grouped by Duration
plt.figure(figsize=(12, 6))
sns.boxplot(x=pd.cut(df['Duration'], bins=[0, 30, 60, 120, 500], labels=['<30', '30-60', '60-120', '>120']),
            y='Maxpulse' , hue='Duration_Category' , data=df, palette='Set3' , dodge=False,legend=False )
plt.title('Maxpulse by Duration Category')
plt.xlabel('Duration Category (minutes)')
plt.ylabel('Maxpulse')
plt.show()
```

```python
# 4. Bar Plot: Average Maxpulse by Duration Category
# Create Duration categories to ensure the column exists
df['Duration_Category'] = pd.cut(df['Duration'], bins=[0, 30, 60, 120, 500], labels=['<30', '30-60', '60-120', '>120'])
plt.figure(figsize=(10, 6))
sns.barplot(x='Duration_Category', y='Maxpulse',  hue='Duration_Category', data=df, palette='Set2', estimator='mean' , dodge=False, legend=False)
plt.title('Average Maxpulse by Duration Category')
plt.xlabel('Duration Category (minutes)')
plt.ylabel('Average Maxpulse')
plt.show()
```

```python
# 5. Heatmap: Correlation Matrix
# Compute the correlation matrix
correlation_matrix = df[['Duration', 'Pulse', 'Maxpulse', 'Calories']].corr()

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```
